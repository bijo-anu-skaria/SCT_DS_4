```python
#libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load datasets
road_weather=pd.read_csv("/content/Acc_Classified_according_to_Type_of_Weather_Condition_2014_and_2016.csv")
road_cond=pd.read_csv("/content/Acc_clf_acco_to_Road_Cond_2014_and_2016.csv")
```

```python
#combine both datasets into single one
road_acc=pd.concat([road_weather,road_cond])
road_acc.head()
```

```python
road_acc.info()
```

```python
road_acc.describe()
```

```python
road_acc.isnull().sum()
```

```python
#handle missing values
road_acc.fillna(road_acc.mean(numeric_only=True), inplace=True)
road_acc.isnull().sum()
```

```python
road_acc.shape
```

```python
#outlier checking
numeric_cols = road_acc.select_dtypes(include=np.number)
Q1=numeric_cols.quantile(0.25)
Q3=numeric_cols.quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
print(lower_bound)
print(upper_bound)
print(IQR)
```

```python
#boxplot
plt.figure(figsize=(10,6))
sns.boxplot(data=road_acc)
plt.show()
```

```python
#remove outliers
outlier_indices = road_acc[(numeric_cols < lower_bound) | (numeric_cols > upper_bound)].dropna(how='all').index
road_acc = road_acc.drop(outlier_indices)
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=road_acc)
plt.title("Boxplot after removing outliers")

plt.show()
```

```python
road_acc.shape
```

```python
# Calculate the correlation matrix for numerical columns
correlation_matrix = road_acc.select_dtypes(include=np.number).corr()

# Display the correlation matrix (optional, can be large)
# display(correlation_matrix)

# Visualize the correlation matrix as a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=False) # annot=False to avoid overcrowding with many columns
plt.title('Correlation Matrix of Road Accident Data')
plt.show()
```

The heatmap visually represents the correlation between all pairs of numerical columns in the dataset. Red colors indicate strong positive correlations, blue colors indicate strong negative correlations, and colors close to white or gray indicate weak correlations. This can help identify potential relationships between different factors like weather conditions, road conditions, and accident counts.

```python
# Visualize accident hotspots by State/UT
plt.figure(figsize=(15, 8))
state_accident_counts = road_acc.groupby('State/ UT')['Fine - Total Acc. - 2014'].sum().sort_values(ascending=False) # Using 'Fine - Total Acc. - 2014' as an example, can be expanded to other columns
sns.barplot(x=state_accident_counts.index, y=state_accident_counts.values)
plt.xticks(rotation=90)
plt.title('Total Accidents by State/UT (Example: Fine Weather 2014)')
plt.ylabel('Total Accidents')
plt.xlabel('State/UT')
plt.tight_layout()
plt.show()

# Visualize accident counts by weather conditions (example for 2014)
weather_cols_2014 = [col for col in road_acc.columns if 'Total Acc. - 2014' in col and any(weather_type in col for weather_type in ['Fine', 'Mist/fog', 'Cloudy', 'Light rain', 'Heavy rain', 'Flooding', 'Hail/sleet', 'snow', 'Strong wind', 'Dust storm', 'Very hot', 'Very cold', 'Other extraordinary weather condition'])]
weather_accident_counts_2014 = road_acc[weather_cols_2014].sum().sort_values(ascending=False)

plt.figure(figsize=(15, 8))
sns.barplot(x=weather_accident_counts_2014.index, y=weather_accident_counts_2014.values)
plt.xticks(rotation=90)
plt.title('Total Accidents by Weather Condition (2014)')
plt.ylabel('Total Accidents')
plt.xlabel('Weather Condition')
plt.tight_layout()
plt.show()

# Visualize accident counts by road conditions (example for 2016)
road_cols_2016 = [col for col in road_acc.columns if 'Number of Accidents - 2016' in col or 'Road-Accident - 2016' in col]
road_accident_counts_2016 = road_acc[road_cols_2016].sum().sort_values(ascending=False)

plt.figure(figsize=(15, 8))
sns.barplot(x=road_accident_counts_2016.index, y=road_accident_counts_2016.values)
plt.xticks(rotation=90)
plt.title('Total Accidents by Road Condition (2016)')
plt.ylabel('Total Accidents')
plt.xlabel('Road Condition')
plt.tight_layout()
plt.show()
```

Certain states like Delhi, Chhattisgarh, Orissa, Punjab, Assam, and Goa appear to be accident hotspots. Further investigation into the specific conditions and factors contributing to accidents in these states would be beneficial.
Regarding road conditions in 2016, normal surfaced roads account for the highest number of accidents, followed by the 'Others' category and Kutcha roads. This highlights the need to understand what constitutes the 'Others' category and to focus on improving road safety on all types of roads.


```python
# Visualize accident counts by weather conditions (example for 2014)
weather_cols_2014 = [col for col in road_acc.columns if 'Total Acc. - 2014' in col and any(weather_type in col for weather_type in ['Fine', 'Mist/fog', 'Cloudy', 'Light rain', 'Heavy rain', 'Flooding', 'Hail/sleet', 'snow', 'Strong wind', 'Dust storm', 'Very hot', 'Very cold', 'Other extraordinary weather condition'])]
weather_accident_counts_2014 = road_acc[weather_cols_2014].sum().sort_values(ascending=False)

plt.figure(figsize=(15, 8))
sns.barplot(x=weather_accident_counts_2014.index, y=weather_accident_counts_2014.values)
plt.xticks(rotation=90)
plt.title('Total Accidents by Weather Condition (2014)')
plt.ylabel('Total Accidents')
plt.xlabel('Weather Condition')
plt.tight_layout()
plt.show()
```

In conclusion, road accidents are influenced by a combination of factors including location (State/UT), weather conditions, and road conditions. While fine weather accounts for the highest number of accidents in 2014, other extraordinary weather conditions and different road types also play a significant role. Further investigation into the specific reasons for high accident rates in the identified hotspots and under the significant contributing factors would be beneficial for developing targeted accident prevention strategies.
